```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


df = pd.read_csv("Customer Churn.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ","0")
#converting the total charges object to float
```

```python
df['TotalCharges'] = df['TotalCharges'].astype("float")
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df['customerID'].duplicated().sum() 
```

```python
#making function to change the values of a column

def conv(value):
  if value == 1:
    return 'yes'
  else:
    return 'no'
  
  
```

```python
#changing the value of the SeniorCitizen to yes and no
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
#plotting the yes and no values in the churn column
churnCount = sns.countplot(x = df["Churn"] , data = df )
plt.title("count of customers by churn", fontsize= 20
)
churnCount.bar_label(churnCount.containers[0]) 
```

```python
#checking tha same data in pie chart percentage
plt.figure(figsize= (5,5))
gb = df.groupby('Churn').agg({'Churn':"count"})
gb 

plt.pie(gb['Churn'], labels = {'Yes','No'} , autopct = "%1.2f%%")
plt.title("percentage of the churn customers to non churn", fontsize = 10)

plt.show()

```

```python
#from the above pie chart 26.54 percent customers have churned out
#let's find the reason behind it according to the gender.
plt.figure(figsize = (4,4))
plt.title("Churn by gender")
sns.countplot(x = "gender", data= df, hue = "Churn")
```

```python
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
```

```python


counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

ax = percentages.plot(kind='bar', stacked=True, figsize=(8,6), colormap='Set2')


for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

plt.xlabel('Senior Citizen')
plt.ylabel('Percentage')
plt.title('Churn Distribution by Senior Citizen Status')
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
#From the results below we can check that 41% of the churn is for senior citizen and only 23 % for others
```
