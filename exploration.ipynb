```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


df = pd.read_csv("Customer Churn.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ","0")
#converting the total charges object to float
```

```python
df['TotalCharges'] = df['TotalCharges'].astype("float")
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df['customerID'].duplicated().sum() 
```

```python
#making function to change the values of a column

def conv(value):
  if value == 1:
    return 'yes'
  else:
    return 'no'
  
  
```

```python
#changing the value of the SeniorCitizen to yes and no
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
#plotting the yes and no values in the churn column
churnCount = sns.countplot(x = df["Churn"] , data = df )
plt.title("count of customers by churn", fontsize= 20
)
churnCount.bar_label(churnCount.containers[0]) 
```

```python
#checking tha same data in pie chart percentage
plt.figure(figsize= (5,5))
gb = df.groupby('Churn').agg({'Churn':"count"})
gb 

plt.pie(gb['Churn'], labels = {'Yes','No'} , autopct = "%1.2f%%")
plt.title("percentage of the churn customers to non churn", fontsize = 10)

plt.show()

```

```python
#from the above pie chart 26.54 percent customers have churned out
#let's find the reason behind it according to the gender.
plt.figure(figsize = (4,4))
plt.title("Churn by gender")
sns.countplot(x = "gender", data= df, hue = "Churn")
```

```python
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
```

```python

counts =df.groupby(["SeniorCitizen","Churn"]).size().unstack()
percentage = counts.div(counts.sum(axis=1),axis=0)*100

stacking = percentage.plot(kind = "bar", stacked = True, figsize=(6,5),colormap='Set2')

for container in stacking.containers:
  stacking.bar_label(container,fmt='%.1f%%', label_type="center")
plt.xlabel('senior citizen')
plt.ylabel('percentage')
plt.xticks(rotation=0) #this rotates the xlable 
plt.title('churn distribution by senior citizens')
plt.legend(title="churn")
plt.grid(False)  # disables all gridlines
plt.tight_layout()
plt.show() 

#From the results below we can check that 41% of the churn is for senior citizen and only 23 % for others
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x="tenure" , data = df, bins = df["tenure"].max(), # we can use the max function to get the bin size, instead manually checking it.
             hue = "Churn")
plt.show()
```

```python

plt.figure(figsize = (4,4))
churnData = sns.countplot(x = "Contract" , data = df, hue = "Churn")
churnData.bar_label(churnData.containers[0])
plt.title("count of customers by Contract")
plt.show()

# the below chart concludes that people who have the contract for longer peroid have low percentage of churn.
```

```python
#now in the above graph the 0 index is labeled with number and 1 index is not, now we will try to label both.
plt.figure(figsize=(5,5))
ChurnData = sns.countplot(x = 'Contract',data=df,hue="Churn")

for container in churnData.containers:
  ChurnData.bar_label(container)

plt.title("count of customers by contract")
plt.show()
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
           'StreamingMovies']

n_cols =3
n_rows = (len(columns)+n_cols-1)//n_cols
fig,axes = plt.subplots(n_rows,n_cols,figsize=(15,n_rows * 4))

axes = axes.flatten()

for i,col in enumerate(columns):

  sns.countplot(x=col, data = df , ax=axes[i],hue=df["Churn"])

  axes[i].set_title(f'count plot of {col}')
  axes[i].set_xlabel(col) #added by seaborn by default, incase you want to override use this
  axes[i].set_ylabel('Count') # same as above

for j in range(i + 1, len(axes)):
   fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
